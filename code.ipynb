```python
import pandas as pd 
import datetime
import numpy as np
```

```python
#Convert default excel to csv format
df = pd.read_excel("./dari-mesin.xlsx")
df.to_csv("./dari-mesin.csv", sep=",")

#Read CSV file
df = pd.read_csv("dari-mesin.csv") 
del df['Unnamed: 0']
```

```python
#Change string to datetime format
df['Time']=pd.to_datetime(df.Time)

#Split datetime to date & time
df['new_date'] = [d.date() for d in df['Time']]
df['new_time'] = [d.time() for d in df['Time']]

#Convert 24Hour - 12 Hour

datelist = []
for i in df['new_time']:
    format1 = "%H:%M:%S"
    date = datetime.datetime.strptime(str(i), format1)
    date = date.strftime("%-I:%M %p")
    datelist.append(date)
    
df['new_time'] = pd.DataFrame({'new_time':datelist})

#Remove existing Time col
df = df.drop(columns=['Time'])
```

```python
#Get unique dates & names
dates = np.unique(df['new_date'])
names = np.unique(df['Name'])

#Change all column to str
df = df.applymap(str)

#Define new df
dfbaru = pd.DataFrame()
```

```python
global validco, validci

cols = ['name', 'staffid', 'clock-in', 'clock-out', 'remarks', 'date']
lst = []

for c, date in enumerate(dates):
    date = str(date)
    print("\n [{}] Date :{} \n".format(c, date))
    file_name = "attendancenewww_{}.csv".format(str(date))
    lst = []
    for count, name in enumerate(names):
        row = df.loc[(df['Name'] == name) & (df['new_date'] == str(date))]
        
        staffid = np.array(row['AC-No.'])
        staffid = set(staffid)
        staffid = ''.join(staffid)
        
        print("{}. ({} - {}) and date {}".format(count, name, staffid, date))

        time = np.array(row['new_time'])
        times = time.tolist()
        print(times)
        
        validco, validci = False, False
        
        for time in times :
            
            if "AM" in time:
    
                if validci == False :
                    print("Clock in : {}".format(time))
                    clockin = time
                    validci = True
                
                else : 
                    clockin = str(0)
                    validci = True
                    
            elif "PM" in time:
            
                if validco == False :
                    print("Clock out : {}".format(time))
                    clockout = time
                    validco = True
                
                else : 
                    clockout = str(0)
                    validco = True
                    
            else :
                validci = False
                validco = False
            
            print("Valid CI : {}".format(validci))
            print("Valid CO : {}".format(validco))
            
        if (validci == True) & (validco == True):
            print("Remarks {}\n ========\n".format("VALID!"))
            remarks = "Valid"

        elif (validci == True) | (validco == True):
            print("Remarks {}\n ========\n".format("NON-VALID"))
            remarks = "Non Valid"
            
        elif (validci == False) & (validco == False):
            print("Remarks {}\n ========\n".format("MC"))
            remarks = "-"
            clockin = str(0)
            clockout = str(0)
        
        print("{}, {}, {}, {}, {}, {}".format(name, staffid, clockin, clockout, remarks, str(date)))
        lst.append([name, staffid, clockin, clockout, remarks, str(date)])
        

        df1 = pd.DataFrame(lst, columns=cols)
        df1.to_csv(file_name,sep = ',',index=False, header=True)

    
            
```
